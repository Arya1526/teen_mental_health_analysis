```python
import pandas as pd

file_path = "/Users/Aryaanil/Documents/Teen_Mental_Health_Dataset.csv"

df = pd.read_csv(file_path)

numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

for col in numeric_columns:
    df[col] = df[col].fillna(df[col].mean())

for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Remaining Missing Values:")
print(df.isnull().sum())

print("\nTotal Remaining Missing Values:")
print(df.isnull().sum().sum())
```

```python
import pandas as pd

file_path = "/Users/Aryaanil/Documents/Teen_Mental_Health_Dataset.csv"

df = pd.read_csv(file_path)

df = df[(df["age"] >= 10) & (df["age"] <= 19)]

df = df[(df["sleep_hours"] >= 0) & (df["sleep_hours"] <= 24)]


numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_columns:
    df = df[df[col] >= 0]

print("Cleaned Dataset Shape:")
print(df.shape)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

file_path = "/Users/Aryaanil/Documents/Teen_Mental_Health_Dataset.csv"

df = pd.read_csv(file_path)

columns = ["stress_level", "anxiety_level", "sleep_hours"]

standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

standard_scaled = standard_scaler.fit_transform(df[columns])

minmax_scaled = minmax_scaler.fit_transform(df[columns])

standard_scaled_df = pd.DataFrame(
    standard_scaled,
    columns=["stress_standard", "anxiety_standard", "sleep_standard"]
)

minmax_scaled_df = pd.DataFrame(
    minmax_scaled,
    columns=["stress_minmax", "anxiety_minmax", "sleep_minmax"]
)

comparison_df = pd.concat(
    [df[columns], standard_scaled_df, minmax_scaled_df],
    axis=1
)

print(comparison_df.head())
```

```python
import pandas as pd

file_path = "/Users/Aryaanil/Documents/Teen_Mental_Health_Dataset.csv"

df = pd.read_csv(file_path)

df["Stress_Category"] = df["stress_level"].apply(
    lambda x: "High" if x > 8 else ("Medium" if x >= 5 else "Low")
)

df["Sleep_Category"] = df["sleep_hours"].apply(
    lambda x: "Good Sleep" if x >= 7 else ("Average Sleep" if x >= 5 else "Poor Sleep")
)

print("Stress Category Counts:")
print(df["Stress_Category"].value_counts())

print("\nSleep Category Counts:")
print(df["Sleep_Category"].value_counts())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

file_path = "/Users/Aryaanil/Documents/Teen_Mental_Health_Dataset.csv"

df = pd.read_csv(file_path)

X = df.drop("depression_label", axis=1)

X = pd.get_dummies(X)

y = df["depression_label"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("X_train Shape:")
print(X_train.shape)

print("\nX_test Shape:")
print(X_test.shape)

print("\ny_train Shape:")
print(y_train.shape)

print("\ny_test Shape:")
print(y_test.shape)
```
